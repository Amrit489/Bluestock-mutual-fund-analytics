```python
import pandas as pd
```

```python
transactions = pd.read_csv("../Data/Raw/08_investor_transactions.csv")

print(transactions.shape)

transactions.head()
```

```python
transactions.info()
```

```python
transactions.isnull().sum()
```

```python
transactions["transaction_type"].value_counts()
```

```python
invalid_amounts = transactions[
    transactions["amount_inr"] <= 0
]

print("Invalid Amount Records:", len(invalid_amounts))
```

```python
transactions["kyc_status"].value_counts()
```

```python
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

transactions.info()
```

```python
transactions.to_csv(
    "../Data/Processed/clean_transactions.csv",
    index=False
)

print("clean_transactions.csv saved successfully!")
```

```python
performance = pd.read_csv(
    "../Data/Raw/07_scheme_performance.csv"
)

print(performance.shape)

performance.head()
```

```python
performance.info()
```

```python
performance.isnull().sum()
```

```python
negative_sharpe = performance[
    performance["sharpe_ratio"] < 0
]

print("Negative Sharpe Funds:", len(negative_sharpe))
```

```python
invalid_expense = performance[
    (performance["expense_ratio_pct"] < 0.1) |
    (performance["expense_ratio_pct"] > 2.5)
]

print("Invalid Expense Ratios:", len(invalid_expense))
```

```python
performance.to_csv(
    "../Data/Processed/clean_performance.csv",
    index=False
)

print("clean_performance.csv saved successfully!")
```
